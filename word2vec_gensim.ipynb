```python
import pandas as pd
import jieba
from gensim.models.word2vec import Word2Vec
import numpy as np

# 读入训练集文件
data = pd.read_csv('train.csv')
# 转字符串数组
corpus = data['comment'].values.astype(str)
# 分词，再重组为字符串数组
corpus = [jieba.lcut(corpus[index]
                      .replace("，", "")
                      .replace("!", "")
                      .replace("！", "")
                      .replace("。", "")
                      .replace("~", "")
                      .replace("；", "")
                      .replace("？", "")
                      .replace("?", "")
                      .replace("【", "")
                      .replace("】", "")
                      .replace("#", "")
                    ) for index in range(len(corpus))]
# 初始化 Skip-Gram 模型
model = Word2Vec(
    corpus,
    sg=1,  # 使用 Skip-Gram 训练
    vector_size=300,  # 词向量维度
    window=5,       # 上下文窗口大小
    min_count=3,     # 忽略低频词的最小次数
    workers=4        # 工作者数量（并行训练）
)
```

```python
#模型显示
print('模型参数：',model,'\n')
```

```python
print('最匹配的词是：',model.wv.most_similar(positive=['好吃'], negative=['难吃']),'\n')
```

```python
env_vector = model.wv.__getitem__('环境')

# 打印词向量及其形状
print('词向量：', env_vector)
```

```python
restaurant_vector = model.wv["餐厅"]
party_vector = model.wv["聚会"]
quiet_vector = model.wv["安静"]
result = restaurant_vector+party_vector-quiet_vector
max_similarity = -1.0
best_word = ""

for word in model.wv.index_to_key:  # 遍历所有单词
    vec = model.wv[word]            # 获取对应向量
    similarity = np.dot(result, vec) / (np.linalg.norm(result) * np.linalg.norm(vec))
    if similarity > max_similarity:
        max_similarity = similarity
        best_word = word

print(f"最相关的词是：{best_word}")
```
